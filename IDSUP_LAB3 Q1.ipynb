```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
import math
```

```python
data = [i for i in range(60,70)]
[data.append(random.choice(data)) for i in range(1000)]
print(data)
```

```python
df = pd.Series(data)
df
```

```python
mean_val = df.mean()
print(f'Mean : {mean_val}')
```

```python
std_val = df.std()
print(f'Standard deviation : {std_val}')
```

```python
frq = collections.Counter(data)
plt.hist(data)

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Values between 60 and 70')
plt.show()
```

```python
sample = []
smean = []


for i in range(30):
    for j in range(30):
        val = random.choice(data)
        sample.append(val)
    temp_mean = np.mean(sample)
    smean.append(temp_mean)
print(smean)
```

```python
for i in range(470):
    for j in range(20):
        val = random.choice(data)
        sample.append(val)
    temp_mean = np.mean(sample)
    smean.append(temp_mean)
print(smean)
```

```python
plt.hist(smean)

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
print(f'mean of Sample mean = {np.mean(smean)}')
```

```python
print(f'Standard deviation = {np.std(smean)}')
```
